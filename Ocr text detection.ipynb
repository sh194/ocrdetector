## Now lets import the required libraries.

```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Reading Image

## I already have few images from which I am reading one of them.

```python
img_path="sign1.png"
```

## Printing image

### the image which I read is shown below as we can see.

```python
img = cv2.imread(img_path)
plt.imshow(img)
plt.show()
```

## Processing image

```python
reader = easyocr.Reader(['en'],gpu=False)
result = reader.readtext(img_path)
result[0]
```

## Printing results

### Our results are ready lets view one line of it.

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(img_path)
img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5)
img = cv2.putText(img, text, top_left, font, 5, (255,0,0), 4)
plt.figure(figsize = (10,10))
plt.imshow(img)
plt.show()
```

### here we can see ractangular box is not vissible nither the characters so lets change some of the parameters and run it again.

### the rectangular box shows the detected image and the text detected is on the top corner of the rectangle.

## Multiple lines...?

#### Making changes for the multiple lines too.

```python
img = cv2.imread(img_path)
for line in result:
    t_left = tuple([int(x) for x in line[0][0]])
    b_right = tuple([int(x) for x in line[0][2]])
    text = line[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    img = cv2.rectangle(img, t_left, b_right, (0,255,0), 5)
    img = cv2.putText(img, text, t_left, font, 5, (255,0,0), 4)
    
plt.figure(figsize=(15,15))
plt.imshow(img)
plt.show()
```
